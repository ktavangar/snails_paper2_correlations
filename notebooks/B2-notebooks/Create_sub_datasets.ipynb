```python
import numpy as np
import pandas as pd
import astropy.table as at
import pickle
import glob
```

```python
directory = '../../data/mssa_channels_B2/'

for file in glob.glob(directory+'*'):
    filedata = np.loadtxt(file)
    file_data_no_time = filedata[:,1:]
    
    outer_data = file_data_no_time[:,240:]
    inner_data = file_data_no_time[:,:256]

    np.savetxt(file[:-11]+'outer.dat',
           np.hstack([np.reshape(filedata[:,0], (len(filedata[:,0]), 1)), outer_data]))
    np.savetxt(file[:-11]+'inner.dat',
           np.hstack([np.reshape(filedata[:,0], (len(filedata[:,0]), 1)), inner_data]))
```
