## Plotting Sgr Orbit

```python
dat_ = pd.read_csv('../data/S1-orbit.txt', sep=' ', names=['x', 'y', 'z', 'Rxy', 'R', 'vx', 'vy', 'vz', 'A', 'B'])
dat=dat_[(dat_.index<6) & (dat_.index>2.)]
```

```python
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(12,5))
ax1.plot(dat.index,dat.x,label='x (kpc)',color='blue')
ax1.plot(dat.index,dat.y,label='y (kpc)',color='red')
ax1.plot(dat.index,dat.z,label='z (kpc)',color='green')
# ax1.axhline(sgr_orb1.x(time1[cross_idx1])*ro, c='k', ls='--')
# ax1.axvline(time1[cross_idx1]*conversion.time_in_Gyr(vo,ro), c='k', ls='--')
ax1.set_xlabel(r'$\mathrm{Time\ (Gyr)}$')
ax1.set_ylabel(r'$x,y,z\ (\mathrm{kpc})$')
ax1.axhline(0)
ax1.set_xlim(2,6)
ax1.set_ylim(-100, 100)

im2=ax2.scatter(dat.x, dat.z, c=dat.index)
plt.colorbar(im2)
ax1.legend()
plt.show()



```

```python
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(12,5))
ax1.plot(dat.index,np.sqrt(dat.x**2+dat.y**2+dat.z**2),label='r (kpc)',color='blue')
# ax1.plot(dat.index,dat.y,label='y (kpc)',color='red')
# ax1.plot(dat.index,dat.z,label='z (kpc)',color='green')
# ax1.axhline(sgr_orb1.x(time1[cross_idx1])*ro, c='k', ls='--')
# ax1.axvline(time1[cross_idx1]*conversion.time_in_Gyr(vo,ro), c='k', ls='--')
ax1.set_xlabel(r'$\mathrm{Time\ (Gyr)}$')
ax1.set_ylabel(r'$x,y,z\ (\mathrm{kpc})$')
ax1.axhline(20)
ax1.set_xlim(2,7)
# ax1.set_ylim(-100, 100)

im2=ax2.scatter(dat.x, dat.z, c=dat.index)
plt.colorbar(im2)
ax1.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1,4,figsize=(16,4))

axs[0].plot(time*conversion.time_in_Gyr(220.,8.),8.*sgrvas.x(time), label=r'$X$ (kpc)', color='blue')
axs[0].plot(time*conversion.time_in_Gyr(220.,8.),8.*sgrvas.z(time), label=r'$Z$ (kpc)', color='green')
axs[0].set_xlabel(r'$\mathrm{Time\ (Gyr)}$')
axs[0].set_ylabel(r'$x,z\ (\mathrm{kpc})$', labelpad=-10)
axs[0].axhline(0, ls='--', c='k')
axs[0].set_xlim(0,3)
axs[0].set_ylim(-100, 100)
axs[0].legend(loc='lower right', frameon=True, framealpha=1)


im2=axs[1].scatter(8.*sgrvas.x(time), 8.*sgrvas.z(time), c=time*conversion.time_in_Gyr(220.,8.), vmin=0.1, vmax=0.7)
axs[1].set_xlim(-100, 70)
axs[1].set_ylim(-100, 100)
axs[1].set_xlabel(r'$X$ [kpc]')
axs[1].set_ylabel(r'$Z$ [kpc]', labelpad=-10)
# axs[1].set_aspect('equal')
axs[1].plot([-15,-5], [0,0], lw=5, c='k', label='Disk region \n explored')
axs[1].plot([5,15], [0,0], lw=5, c='k')
axs[1].legend(loc='upper left')
plt.colorbar(im2, label='Time [Gyr]', pad=0)


axs[2].plot(dat.index,dat.x,label='x (kpc)',color='blue')
axs[2].plot(dat.index,dat.z,label='z (kpc)',color='green')
axs[2].plot(dat.index,dat.y,label='y (kpc)',color='orange')
axs[2].set_xlabel(r'$\mathrm{Time\ (Gyr)}$')
axs[2].set_ylabel(r'$x,z\ (\mathrm{kpc})$', labelpad=-10)
axs[2].axhline(0, ls='--', c='k')
axs[2].set_xlim(2,6)
axs[2].set_ylim(-100, 100)

im4=axs[3].scatter(dat.x, dat.z, c=dat.index, vmin=0, vmax=6)
axs[3].set_xlabel(r'$X$ [kpc]')
axs[3].set_ylabel(r'$Z$ [kpc]', labelpad=-10)
axs[3].set_xlim(-100, 70)
axs[3].set_ylim(-100, 100)
axs[3].plot([-20,-5], [0,0], lw=5, c='k', label='Disk region \n explored')
axs[3].plot([5,20], [0,0], lw=5, c='k')
# axs[3].set_aspect('equal')
plt.colorbar(im4, label='Time [Gyr]', pad=0)



fig.text(0.25, 0.95, 'Test Particle Simulation Orbit', ha='center', fontsize=16)
fig.text(0.75, 0.95, 'N-body Simulation Orbit', ha='center', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('../figures/sim_orbits.pdf')
plt.show()
```
