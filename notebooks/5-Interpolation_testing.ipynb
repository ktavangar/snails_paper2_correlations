**My thoughts on this:**

I can see why it could be a problem and not super rigorous. But I still think it would be valuable to do as an addition to the analysis. It is the only way we are going to be able to look at how all the ocefficients change throughout the disk.

One test to see if it is reasonable to do this would be to check how many stars have periods of less than the time difference between each step. The way to check would be to see where the difference between consecutive thetas is more than $\pi/2$. This is still fine for our purposes but an issue would arise if/when there are periods of less than the time difference, which would result in theta appearing to progress much more slowly than in reality. I'm assuming the distribution of theta steps would be approximately exponential (similar to vertical density distribution in the disk). If this would imply a large number of stars for which this hack would not work, then we will need to abandon.

The first-order thing to check would be for the unperturbed case because if it doesn't work there, then there is no chance of it working later.

```python
all_theta_steps = []
all_comb = np.zeros((1,3))
for i in np.unique(t_700_800.index):
    tab = t_700_800_by_name.get_group(i)
    jphi_ = tab['jphi'].values
    theta_steps = (tab['theta_z'][1:] - tab['theta_z'][:-1]).values
    theta_steps[np.where(theta_steps < 0)[0]] = 2*np.pi + theta_steps[np.where(theta_steps < 0)[0]]
    comb = np.vstack((theta_steps, jphi_[:-1], [i]*len(theta_steps))).T
    all_theta_steps = np.append(all_theta_steps, theta_steps)
    all_comb = np.append(all_comb, comb[1:], axis=0)
    
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(12,6))
ax1.hist(all_theta_steps, bins = 20)
ax1.set_yscale('log')

ax2.scatter(all_comb[:,0], all_comb[:,1], c=all_comb[:,2], s=1)
ax2.set_yscale('log')
plt.show()
```

This is telling me that there are many(?) stars that at least occasionally have periods shorter than the timesteps. However, we also know that there are cases where the stars do actually go backwards in phase due to something and then continue as they were before. The ones where the difference is just under $2\pi$ are likely to be those. THe question is how far does that extend and how many are not one-off backwards trends that later correct?

Some thoughts: What I am realizing is that the only scenario we care about when deciding whether or not to implement this interpolation is the one where the period of the z "orbit" is less than the timestep. In the other scenarios, namely when there are departures from the smoothness of the $\theta_z$ curve, that is just something that will happen with any discrete data set. We would just need to make explicit that we did this interpolation and therefore our simulation is not fully self-consistent on timescales less than the time between timesteps. But this seems fine to me as the time resolution is always a factor in analyses such as this; we would just have a time resolution not shown by what we are presenting. 

Therefore, I think the takeaway from these plots is that we can do the interpolation. What I was worried about is a trend of increasing $\Delta \theta$ with decreasing $J_{\phi}$ and whether that implied that some $\Delta \theta > 2\pi$ for some stars. In that scenario, we would be unable to interpolate correctly. However, based on these plots, that doesn't appear to be the case; there don't seem to be any (or $<1\%$) stars with $\Delta \theta > 5.5$, even in the completely perturbed disk. This was of course using a small subset, and I should see what happens with all the stars.
